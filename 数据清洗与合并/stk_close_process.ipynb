```python
import pandas as pd 
import json
df_pipe = pd.read_excel('/Users/vitol/vscode/firm_perf/firm_perf_code/最终数据/Stk_info_industry_total_cleaned.xlsx',index_col=False)
df_pipe.drop(columns=['Unnamed: 0'],inplace=True)
df_pipe = df_pipe.rename(columns = {'Stkcd_old':'Stkcd'})
df_close = pd.read_csv('/Users/vitol/vscode/firm_perf/firm_perf_code/最终数据/股票收盘价_涨跌幅_换手率.csv')
df_close = df_close.rename(columns = {'CLOSE':'收盘价','CLOSE_pct_change':'涨跌幅','TURN':'换手率'})
df_close.head()
```

```python
df_close = pd.merge(df_close, df_pipe, on=['Stkcd'],how = 'left')
df_basic_info = pd.read_excel('/Users/vitol/vscode/firm_perf/firm_perf_code/最终数据/股票基本信息（含行业）drop后.xlsx')[['stk_code','Stknme','公司全称']]
df_basic_info = df_basic_info.rename(columns = {'stk_code':'Stkcd_filled'})
df_close = pd.merge(df_close, df_basic_info, on=['Stkcd_filled'],how = 'left')
df_close
```

```python
df_close['id'] = df_close.apply(lambda row: f"{row['Stkcd']} {row['Stknme']} {row['公司全称']}", axis=1)
df_close = df_close.drop(columns=['Stkcd','year','申万行业分类','Stkcd_filled','Stknme', '公司全称'])
df_close
```

```python
json_list = []
col = df_close.columns.tolist()

def trans(group):
    content = {}
    for _, row in group.iterrows():
        for i in col:
            if i != 'id' and i != 'date':
                content[f'{row["date"]}这一天的{i}'] = row[i]
    text = {
        'text': content,
        "id": row['id']
    }
    json_list.append(text)

df_close.groupby('id').apply(trans)
```

```python
# 指定要保存的文件名
filename = 'stk_close.json'

# 使用 with 语句打开文件，并使用 json.dump 方法将 json_list 写入文件
with open(filename, 'w', encoding='utf-8') as f:
    # 如果 json_list 是一个列表或字典，直接使用 dump
    json.dump(json_list, f, ensure_ascii=False, indent=4)

print(f"数据已保存到 {filename}")
```
