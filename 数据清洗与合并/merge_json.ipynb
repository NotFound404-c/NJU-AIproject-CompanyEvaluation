```python
import json
import tqdm


with open('/Users/vitol/vscode/firm_perf/firm_perf_code/json文件汇总/stk_basic_info.json', 'r', encoding='utf-8') as file:

    stk_basic_info = json.load(file)


with open('/Users/vitol/vscode/firm_perf/firm_perf_code/json文件汇总/stk_close.json', 'r', encoding='utf-8') as file:

    stk_close = json.load(file)

with open('/Users/vitol/vscode/firm_perf/firm_perf_code/json文件汇总/stk_fin.json', 'r', encoding='utf-8') as file:

    stk_fin = json.load(file)

with open('/Users/vitol/vscode/firm_perf/firm_perf_code/json文件汇总/stk_mv.json', 'r', encoding='utf-8') as file:

    stk_mv = json.load(file)

with open('/Users/vitol/vscode/firm_perf/firm_perf_code/json文件汇总/stk_perf.json', 'r', encoding='utf-8') as file:

    stk_perf = json.load(file)

with open('/Users/vitol/vscode/firm_perf/firm_perf_code/json文件汇总/stk_comments.json', 'r', encoding='utf-8') as file:

    stk_comments = json.load(file)





for i in tqdm.tqdm(stk_basic_info):
    i['text']['公司简称'] = i['text'].pop('Stknme')
    i['text']['股票代码'] = i['text'].pop('Stkcd_old')

```

```python
def mer_json(json_files1, json_files2):
    merged_json = {}

    # 遍历第一个列表
    for item in tqdm.tqdm(json_files1):
        id = item['id']
        merged_json[id] = item['text']

    # 遍历第二个列表
    for item in tqdm.tqdm(json_files2):
        id = item['id']
        if id in merged_json:
            # 如果 id 已经存在，合并 text 字典
            merged_json[id].update(item['text'])
        else:
            # 如果 id 不存在，直接添加
            merged_json[id] = item['text']

    # 将结果转换为列表
    merged_json = [{"text": value, "id": key} for key, value in merged_json.items()]
    return merged_json
```

```python
merged_part1 = mer_json(stk_basic_info,stk_close)
merged_part2 = mer_json(merged_part1,stk_fin)
merged_part3 = mer_json(merged_part2,stk_mv)
merged_part4 = mer_json(merged_part3,stk_perf)
merged_part5 = mer_json(merged_part4,stk_comments)
```
